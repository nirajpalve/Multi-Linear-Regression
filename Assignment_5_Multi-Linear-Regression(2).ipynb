```python
import pandas as pd           # Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
data=pd.read_csv("50_Startups.csv")  # Loading Data
data
```

```python
data.info()  # To check wheather null value exist or not
```

```python
data_1=data.rename({'R&D Spend':'RD','Administration':'ADM','Marketing Spend':'MKTS'},axis=1)   # Renaming the coloumn Labels
data_1
```

```python
data_1.corr() # Importing correlation matrix
```

```python
sns.set_style()
sns.pairplot(data_1)
```

# Model Building

```python
model=smf.ols("Profit~RD+ADM+MKTS", data=data_1).fit()
```

```python
print(model.tvalues, '\n', model.pvalues) # For finding t and p-Values
```

```python
(model.rsquared,model.rsquared_adj) #R squared values and R squared adjusting values
```

```python
ml_A = smf.ols("Profit~ADM", data = data_1).fit()  # Checking Administration p_value individually
print(ml_A.tvalues, '\n', ml_A.pvalues) 
```

```python
ml_M = smf.ols("Profit~MKTS", data = data_1).fit()  # Checking Marketing p_value individually
print(ml_M.tvalues, '\n', ml_M.pvalues) 
```

```python
ml_AM = smf.ols("Profit~ADM + MKTS", data = data_1).fit()  # Checking both p_value together
print(ml_AM.tvalues, '\n', ml_AM.pvalues) 
```

# Model Validation Technique

```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_R = smf.ols("RD~ADM+MKTS",data=data_1).fit().rsquared
vif_R = 1/(1-rsq_R)

rsq_A = smf.ols("ADM~RD+MKTS",data=data_1).fit().rsquared
vif_A = 1/(1-rsq_A)

rsq_M = smf.ols("MKTS~RD+ADM",data=data_1).fit().rsquared
vif_M = 1/(1-rsq_M)

# Putting the values in Dataframe format
d1={'Variables':['RD','ADM','MKTS'],'VIF':[vif_R,vif_A,vif_M]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
# 2) Residual Analysis
# Test for Normality of Residuals (Q-Q Plot)

qqplot = sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-20000))
```

## Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

## Residual Vs Regressors Graphs

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'ADM',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'MKTS',fig=fig)
plt.show()
```

# Model Deletion Diagnostics

## Detecting Influencers/Outliers

## Cooks Distance

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data_1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```
